```python
#Alikhan Bakytbek FIT, CS&S
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

operationsData = pd.read_csv('operations.csv', sep = ',');

datasetData = pd.read_csv('dataset.csv', sep = ',')
codesData = pd.read_csv('codes.csv', sep = ';')
typesData = pd.read_csv('types.csv', sep = ';')


dfOper = pd.DataFrame(operationsData)
dfDataset = pd.DataFrame(datasetData)
dfCodes = pd.DataFrame(codesData)
dfTypes = pd.DataFrame(typesData)

print(operationsData)



data3 = pd.merge(dfDataset, dfOper)
data2 = data3.merge(dfCodes)
data = data2.merge(dfTypes)

print(data)

data[data['client_id'] > 50000000][data['client_id']<90000000]['client_id'].hist(bins=100, color='black')
plt.show()

data['type'].hist(bins = 100, color = 'black')
plt.show()

data['status'].hist(bins = 100, color = 'black')
plt.show()

data['code'].hist(bins = 50, color = 'yellow')
plt.show()

data['sum'].hist(bins = 50, color = 'red')
plt.show()

```
